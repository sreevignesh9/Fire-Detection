```python
from keras.preprocessing.image import load_img 
from keras.preprocessing.image import img_to_array 
from keras.applications.vgg16 import preprocess_input 


from keras.applications.vgg16 import VGG16 
from keras.models import Model

# clustering and dimension reduction
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# for everything else
import os
import numpy as np
import matplotlib.pyplot as plt
from random import randint
import pandas as pd
import pickle
```

```python
X_train = np.load("xtrain.npy")
y_train = np.load("ytrain.npy")
```

```python
model = VGG16()
model = Model(inputs = model.inputs, outputs = model.layers[-2].output)
```

```python
import cv2
import skimage
from skimage.transform import resize
x = []
img = cv2.imread('../dataset/test/fire/abc169.jpg')
img = skimage.transform.resize(img,(224,224), mode = "constant",anti_aliasing=True)
img = img.reshape(224,224,3) 
img_arr = np.asarray(img)
# img_arr
x.append(img_arr)
x = np.asarray(x)

features = model.predict(x)
# print(features)
```

```python
def extract_features(img, model):
#     # load the image as a 224x224 array
#     img = load_img(file, target_size=(150,150))
#     # convert from 'PIL.Image.Image' to numpy array
#     print(img)

    img = np.array(img) 
# #     # reshape the data for the model reshape(num_of_samples, dim 1, dim 2, channels)
    x = [] 
    img = img.reshape(224,224,3) 
    img_arr = np.asarray(img)
    x.append(img_arr)
    x = np.asarray(x)
#     # prepare image for model
#     imgx = preprocess_input(reshaped_img)
#     # get the feature vector
    try:
        features = model.predict(x)
#         print(features)
    except:
        pass
    return features

data = []
p = r"CHANGE TO A LOCATION TO SAVE FEATURE VECTORS"
for img in X_train:
#     print(img)
    # try to extract the features and update the dictionary
    try:
#         print("hi")
        feat = extract_features(img,model)
        print(feat)
        data.append(feat)
    # if something fails, save the extracted features as a pickle file (optional)
    except:
        with open(p,'wb') as file:
            pickle.dump(data,file)
        break
```

```python
feat = np.array(data)
```

```python
feat = feat.reshape(-1,4096)
```

```python
pca = PCA(n_components=100, random_state=22)
```

```python
pca.fit(feat)
x = pca.transform(feat)
```

```python
kmeans = KMeans(n_clusters=2,n_jobs=-1, random_state=22)
kmeans.fit(x)
```

```python
from sklearn import metrics
print ("Accuracy = ", metrics.accuracy_score(y_train[::-1],kmeans.labels_))
```

```python
plt.scatter(x[:,0], 
            x[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], 
            kmeans.cluster_centers_[:, 1], 
            s=50,                             # Set centroid size
            c='red')                           # Set centroid color
plt.show()
```
